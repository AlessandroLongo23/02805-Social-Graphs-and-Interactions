```python
import random
import networkx as nx
import numpy as np

G = nx.Graph()
G.add_edge(0, 1)
while G.number_of_nodes() < 2500:
    edge_list_flatten = np.array(G.edges()).flatten()
    G.add_edge(G.number_of_nodes(), random.choice(edge_list_flatten))

sizes = [G.degree[i] for i in range(G.number_of_nodes())]
args = {
    "node_size": [np.log(size + 1) * 5 for size in sizes],
    "node_color": "salmon",
    "edge_color": "black",
    "node_shape": "o",
    "linewidths": 0.5,
    "width": 0.25,
    "edgecolors": "black",
    "font_color": "black",
    "with_labels": False,
}
nx.draw(G, **args)
```
