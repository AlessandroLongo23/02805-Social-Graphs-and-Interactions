#### Exercise 16
Write a program to generate a table of lexical diversity scores (i.e. token/type ratios), as we saw in 1.1. Include the full set of Brown Corpus genres (nltk.corpus.brown.categories()). Which genre has the lowest diversity (greatest number of tokens per type)? Is this what you would have expected?

```python
def lexical_diversity(words):
    return len(set(words)) / len(words) if words else 0

data = []
for category in brown.categories():
    for fileid in brown.fileids(category):
        words = brown.words(fileid)
        diversity_score = lexical_diversity(words)
        data.append((category, diversity_score))

data.sort(key=lambda x: x[1], reverse=True)

fileids, diversity_scores = zip(*data)

plt.figure(figsize=(15, 7))
plt.bar(fileids, diversity_scores, color='skyblue')

plt.title('Lexical Diversity Across Brown Corpus Categories', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Lexical Diversity Score', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
```

#### Exercise 17
Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def get_most_common_words(text, n=50, include_stopwords=False):
    """
    This function takes a text as input, tokenizes it, removes stopwords,
    and returns the 50 most frequently occurring words.
    """
    words = word_tokenize(text.lower())

    filtered_words = [word for word in words if word.isalpha()]

    if not include_stopwords:
        stop_words = set(stopwords.words('english'))
        filtered_words = [word for word in filtered_words if word not in stop_words]

    fdist = FreqDist(filtered_words)

    return fdist.most_common(n)


alice_text = gutenberg.raw('carroll-alice.txt')
top_50_words = get_most_common_words(alice_text)
for word, frequency in top_50_words:
    print(f"{word}:{frequency}")
```

#### Exercise 18
Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
from nltk.corpus import stopwords

def get_frequent_bigrams_without_stopwords(text, n=50, include_stopwords=False):
    """
    Find the n most frequent bigrams in a text, excluding bigrams containing stopwords.
    
    Args:
        text: A list of words (tokens)
        n: Number of most frequent bigrams to return (default: 50)
        include_stopwords: include stopwords in bigrams (default: False)
    
    Returns:
        A list of tuples (bigram, frequency)
    """
    
    # Create bigrams from the text
    words = word_tokenize(text.lower())
    text_bigrams = list(bigrams(words))

    # Convert words to lowercase for comparison and filter non-alphabetic tokens
    filtered_bigrams = [
        (word1.lower(), word2.lower()) 
        for word1, word2 in text_bigrams 
        if word1.isalpha() 
        and word2.isalpha()
    ]

    if not include_stopwords:
        # Get English stopwords
        stop_words = set(stopwords.words('english'))

        # Filter out bigrams containing stopwords
        filtered_bigrams = [
            (word1, word2) 
            for word1, word2 in filtered_bigrams 
            if word1 not in stop_words 
            and word2 not in stop_words
        ]
    
    # Count frequencies
    bigram_freq = FreqDist(filtered_bigrams)
    
    # Get the n most common bigrams
    return bigram_freq.most_common(n)


alice_text = gutenberg.raw('carroll-alice.txt')
top_50_words = get_frequent_bigrams_without_stopwords(alice_text)
for word, frequency in top_50_words:
    print(f"{word}:{frequency}")
```

#### Exercise 23: Zipf's Law
Let f(w) be the frequency of a word w in free text. Suppose that all the words of a text are ranked according to their frequency, with the most frequent word first. 
Zipf's law states that the frequency of a word type is inversely proportional to its rank (i.e. f  r = k, for some constant k). For example, the 50th most common word type should occur three times as frequently as the 150th most common word type.

Write a function to process a large text and plot word frequency against word rank using pylab.plot. Do you confirm Zipf's law? (Hint: it helps to use a logarithmic scale). What is going on at the extreme ends of the plotted line?
Generate random text, e.g., using random.choice("abcdefg "), taking care to include the space character. You will need to import random first. Use the string concatenation operator to accumulate characters into a (very) long string. Then tokenize this string, and generate the Zipf plot as before, and compare the two plots. What do you make of Zipf's Law in the light of this?

```python
def plot_zipf_law(text):
    """
    Plots word frequency by rank on a log-log scale to visualize Zipf's Law,
    using the get_most_common_words function.
    """
    # 1. Get a large sample of common words to see the pattern clearly.
    # We include stopwords because they are the most frequent words and essential
    # for demonstrating the law.
    most_common = get_most_common_words(text, n=2000, include_stopwords=True)

    # 2. Prepare the data for plotting
    # The rank is the word's position in the frequency list (1, 2, 3, ...)
    ranks = range(1, len(most_common) + 1)
    # The frequency is the count of each word
    freqs = [count for word, count in most_common]

    # 3. Create the plot
    plt.figure(figsize=(10, 6))
    
    # Plot rank vs. frequency as blue dots
    plt.plot(ranks, freqs, 'b.', markersize=4)

    # Set both axes to a logarithmic scale
    plt.xscale('log')
    plt.yscale('log')

    # Add labels and a title for clarity
    plt.title("Zipf's Law for 'Alice in Wonderland'", fontsize=16)
    plt.xlabel('Word Rank (log scale)', fontsize=12)
    plt.ylabel('Word Frequency (log scale)', fontsize=12)
    plt.grid(True, which="both", ls="--", linewidth=0.5)

    # Show the plot
    plt.show()


alice_text = gutenberg.raw('carroll-alice.txt')
plot_zipf_law(alice_text)
```
